```python
import scipy.io
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import filtfilt

data = scipy.io.loadmat('../sampleEEGdata.mat')
data = data['EEG']
times = eeg['times'][0][0].T
srate = data['srate'][0][0][0][0]

# channel to pick
chan2use = 'o1'

chanlocs = []
for i in data['chanlocs'][0][0]['labels'][0]:
    chanlocs.append(i[0].lower())
chanlocs = np.array(chanlocs)

chanidx = np.where(chanlocs == chan2use)[0][0]

# time window for negative peak
# Mike's choices:
negpeaktime = [50, 110]
pospeaktime = [110, 170]

data = data['data'][0][0][chanidx, :, :]
# compute ERP
erp = np.mean(data, axis=1)

# plot ERP
fig, ax = plt.subplots(figsize=(10,6))
ax.plot(times, erp, 'k', linewidth=1)
ax.set_xlim([-300, 1000])

# plot patches over areas
ylim = ax.get_ylim()
ph = ax.fill_betweenx(ylim, negpeaktime[0], negpeaktime[1], color='y', alpha=0.8)

# move the patches to the background
ax.set_ylim(ylim)

plt.show()
```

```python
# first low-pass filter (windowed sinc function)
lowcut = 15
filttime = np.arange(-.3, .3, 1/srate)
filtkern = np.sin(2 * np.pi * lowcut * filttime) / filttime

# adjust NaN and normalize filter to unit-gain
filtkern = np.nan_to_num(filtkern, nan=np.max(filtkern))
filtkern = filtkern / np.sum(filtkern)

# windowed sinc filter
filtkern = filtkern * np.hanning(len(filttime))

# inspect the filter kernel
fig, ax = plt.subplots(2, 1, figsize=(10, 6))

# Time domain
ax[0].plot(filttime, filtkern, 'k', linewidth=2)
ax[0].set_xlabel('Time (s)')
ax[0].set_title('Time domain')

# Frequency domain
hz = np.linspace(0, srate, len(filtkern))
ax[1].plot(hz, np.abs(np.fft.fft(filtkern)), 'ks-', linewidth=2)
ax[1].set_xlim([0, lowcut * 3])
ax[1].set_xlabel('Frequency (Hz)')
ax[1].set_ylabel('Gain')
ax[1].set_title('Frequency domain')

plt.tight_layout()
plt.show()
```

```python
# now filter the ERP and replot
# apply filter
ferp = filtfilt(filtkern,1,erp);


# plot on top of unfiltered ERP
fig, ax = plt.subplots(figsize=(10,7))
ax.plot(times, erp, 'k', linewidth=1)
ax.plot(times, ferp, 'r', linewidth=0.8)
ax.set_xlim([-300, 1000])

# plot patches over areas
ylim = ax.get_ylim()
ax.fill_betweenx(ylim, negpeaktime[0], negpeaktime[1], color='y', alpha=0.8)
ax.fill_betweenx(ylim, pospeaktime[0], pospeaktime[1], color='g', alpha=0.8)

plt.show()
```

```python
# peak-to-peak voltages and timings

def time_to_id(time2plot):
    # convert time in ms to time in indices
    return np.argmin(np.abs(times - time2plot))

times = np.squeeze(times)
#### first for unfiltered ERP

# find minimum/maximum peak values and peak times
def get_p2p(erp):
    negpeaktimeidxs = [time_to_id(negpeaktime[0]), time_to_id(negpeaktime[1])+1]
    pospeaktimeidxs = [time_to_id(pospeaktime[0]), time_to_id(pospeaktime[1])+1]

    erp_neg = erp[negpeaktimeidxs[0] : negpeaktimeidxs[1]]
    times_neg = times[negpeaktimeidxs[0] : negpeaktimeidxs[1]]

    erp_pos = erp[pospeaktimeidxs[0] : pospeaktimeidxs[1]]
    times_pos = times[pospeaktimeidxs[0] : pospeaktimeidxs[1]]

    # ERP timings

    min_erp_peak = np.min(erp_neg)
    min_erp_time = times_neg[np.argmin(erp_neg)]

    max_erp_peak = np.max(erp_pos)
    max_erp_time = times_pos[np.argmax(erp_pos)]

    # get results (peak-to-peak voltage and latency)
    erpP2P = max_erp_peak - min_erp_peak
    erpP2Plat = max_erp_time - min_erp_time
    
    return (erpP2P, erpP2Plat)
```

```python
# then for low-pass filtered ERP
(erpP2P, erpP2Plat) = get_p2p(erp)
(ferpP2P, ferpP2Plat) = get_p2p(ferp)

print(f'Peak-to-peak on unfiltered ERP: {erpP2P:.2f} muV, {erpP2Plat:.2f} ms span.')
print(f'Peak-to-peak on filtered ERP: {ferpP2P:.2f} muV, {ferpP2Plat:.2f} ms span.')
```
